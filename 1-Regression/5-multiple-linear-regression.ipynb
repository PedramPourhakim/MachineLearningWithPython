```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
```

```python
df = pd.read_csv("./FuelConsumption.csv")
```

```python
df.head()
```

```python
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
```

```python
cdf.head(9)
```

```python
plt.scatter(cdf.ENGINESIZE,cdf.CO2EMISSIONS,color='blue')
plt.xlabel("Engine Size")
plt.ylabel("Emission")
plt.show()
```

```python
msk = np.random.rand(len(df)) < 0.8 #اگر عدد رندومم کوچکتر از 0.8 بود داخلش True میگذاریم
train = cdf[msk]
test = cdf[~msk]
```

### Train Data Distribution

```python
plt.scatter(train.ENGINESIZE,train.CO2EMISSIONS,color = 'blue')
plt.xlabel("Engine Size")
plt.ylabel("Emission")
plt.show()
```

## Multiple Regresion Model

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = train[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']]
y = train[['CO2EMISSIONS']]
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('InterCept : ', regr.intercept_)
```

### Prediction

```python
y_hat = regr.predict(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
x = test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']]
y = test[['CO2EMISSIONS']]
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```
